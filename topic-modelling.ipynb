```python
import re  # Import the regular expressions module

def clean(raw):
    """ 
    Remove hyperlinks and markup from the given text.
    
    This function cleans up common HTML elements and special character encodings
    to make the text more readable.
    """
    
    # Replace hyperlinks (anchor tags) with the text "Link."
    result = re.sub("<[a][^>]*>(.+?)</[a]>", 'Link.', raw)

    # Remove HTML entity for '>' (greater than symbol)
    result = re.sub('&gt;', "", result)

    # Replace HTML entity for apostrophe (') with a normal apostrophe
    result = re.sub('&#x27;', "'", result)

    # Replace HTML entity for double quotes (") with actual double quotes
    result = re.sub('&quot;', '"', result)

    # Replace HTML entity for forward slash (/) with a space
    result = re.sub('&#x2F;', ' ', result)

    # Replace paragraph tag `<p>` with a space (preserves spacing)
    result = re.sub('<p>', ' ', result)

    # Remove closing italic tag `</i>` (not needed)
    result = re.sub('</i>', '', result)

    # Remove another HTML entity for '>' (redundant handling)
    result = re.sub('&#62;', '', result)

    # Replace opening italic tag `<i>` with a space (preserves spacing)
    result = re.sub('<i>', ' ', result)

    # Remove newline characters to keep text in a single line
    result = re.sub("\n", '', result)

    return result  # Return the cleaned-up text
```

```python
def remove_nums(text):
    return re.sub(r'\d+','',text)
```

```python
def unify_whitespaces(text):
    return re.sub(r' +',' ',text)
```

```python
from nltk.corpus import stopwords
from nltk import WordNetLemmatizer
import nltk
nltk.download('stopwords')
from nltk.stem import PorterStemmer

stop_words=set(stopwords.words('english'))

def remove_stopwords(text):
    words= [word.lower() for word in text.split() if word.lower() not in stop_words]
    return ' '.join(words)
```

```python
def replace_hearts_with_PAD(text):
    return re.sub(r"[♥]+", ' **** ' ,text)
```

```python
def deEmojify(x):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'', x)
```

```python
def remove_punctuation(text):
    result = "".join(u for u in text if u not in ("?", ".", ";", ":",  "!",'"',',') )
    return result
```

```python
from nltk.stem import SnowballStemmer
nltk.download('punkt')

def Stemming(text):
    tokens=nltk.word_tokenize(text)
    stemmer=SnowballStemmer('english')
    tokens=[stemmer.stem(token) for token in tokens ]
    return " ".join(tokens)
```

```python
def cleaning(df,review):
    df2=df.copy()
    df2[review]=df['review_text']
    df2[review]=df2[review].apply(clean)
    df2[review]=df2[review].apply(remove_nums)
    df2[review]=df2[review].apply( unify_whitespaces)
    df2[review]=df2[review].apply(replace_hearts_with_PAD)
    df2[review] = df2[review].apply(remove_punctuation)
    df2[review] = df2[review].apply(remove_stopwords)
    df2[review] = df2[review].apply(unify_whitespaces)
    df2[review] = df2[review].apply(Stemming)

    return df2
```

```python
!pip install umap.learn
!pip install hdbscan

import umap
import hdbscan 
from functools import partial


from hyperopt import fmin, tpe, hp, STATUS_OK, space_eval, Trials

def generate_clusters(
    message_embeddings,
    n_neighbors,
    n_components,
    min_cluster_size,
    min_samples=None,
    random_state=None,
):
    umap_embeddings = (
        umap.UMAP(
            n_neighbors=n_neighbors,
            n_components=n_components,
            metric="cosine",
        )
    ).fit_transform(message_embeddings)

    clusters = hdbscan.HDBSCAN(
        min_cluster_size=min_cluster_size,
        min_samples=min_samples,
        metric="euclidean",
        gen_min_span_tree=True,
        cluster_selection_method="eom",
    ).fit(umap_embeddings)

    return clusters
```

```python
def score_clusters(clusters, prob_thresh0ld=0.05):
    cluster_labels = clusters.labels_

    label_count = len(np.unique(cluster_labels))
    total_num = len(clusters.labels_)

    cost = np.count_nonzero(clusters.probabilities_ < prob_thresh0ld) / total_num

    return label_count, cost

```

```python
def objective(params, embeddings, label_lower, label_upper):
    clusters = generate_clusters(
        embeddings,
        n_neighbors=params["n_neighbors"],
        n_components=params["n_components"],
        min_cluster_size=params["min_cluster_size"],
    )

    label_count, cost = score_clusters(clusters, prob_thresh0ld=0.05)

    if (label_count < label_lower) | (label_count > label_upper):
        penalty = 1.0
    else:
        penalty = 0

    loss = cost + penalty

    return {"loss": loss, "label_count": label_count, "status": STATUS_OK}
```

```python
def bayesian_search(embeddings, space, label_lower, label_upper, max_evals=100):

    trials = Trials()
    fmin_objective = partial(
        objective,
        embeddings=embeddings,
        label_lower=label_lower,
        label_upper=label_upper,
    )

    best = fmin(
        fmin_objective,
        space=space,
        algo=tpe.suggest,
        max_evals=max_evals,
        trials=trials,
    )

    best_params = space_eval(space, best)
    print("best:")
    print(best_params)
    print(f"label count: {trials.best_trial['result']['label_count']}")

    best_clusters = generate_clusters(
        embeddings,
        n_neighbors=best_params["n_neighbors"],
        n_components=best_params["n_components"],
        min_cluster_size=best_params["min_cluster_size"],
    )

    return best_params, best_clusters, trials
```

```python
from collections import Counter
def compute_IDF(documents):
    word_count=Counter()
    for doc in documents:
        words=set(doc.split())
        word_count.update(words)
    total_words=sum(word_count.values())
    IDF={ word: round(np.log2(total_words/freq)) for word,freq in word_count.items()}

    return IDF
```

```python
def get_group(df,category_col,category):
    single_category=df[df[category_col]==category].reset_index(drop=True)
    return single_category

def most_common(words,n_words):
    word_freq=Counter(words)

    for word in word_freq:
        if word_freq[word]==1:
            pass
        else:
            word_freq[word]*=IDF[word]
    return sorted(word_freq.items(),key=lambda x: -x[1])[:n_words]

```

```python
def extract_label(docs):

    verb=''
    noun1=''
    noun2=''
    adj=''
    dobj=''

    verbs=[]
    dobjs=[]
    nouns=[]
    adjs=[]

    for doc in docs:
        tokens=nlp(doc)
        for token in tokens:
            if (token.is_stop == False ) and (len(str(token).strip()) >0 ):
                if token.pos_ =="VERB":
                    verbs.extend([token.lemma_.lower()])
                elif token.pos_=="NOUN":
                    nouns.extend([token.lemma_.lower()])
                elif token.pos_=="ADJ":
                    adjs.extend([token.lemma_.lower()])
                elif token.dep_=="dobj":
                    dobjs.extend([token.lemma_.lower()])

    def most_common(lst, n_words=1):
        return Counter(lst).most_common(n_words)

    if len(verbs)>0:
        verb=most_common(verbs,1)[0][0]

    if len(dobjs)>0:
        dobj=most_common(dobjs,1)[0][0]

    if len(nouns)>0:
        noun1=most_common(nouns,1)[0][0]
        
    if len(adjs)>1:
        noun2=most_common(nouns,2)[1][0]

    
    label=" ".join([verb,dobj,noun1,noun2])
    return label
    
```

```python
def apply_and_summarize_labels(df,label_col):

    unique_labels=np.unique(df[label_col])
    label_summary={
        label: extract_label(df[df[label_col]==label]['text'].tolist()) for label in unique_labels
    }

    df['summary']=df[label_col].map(label_summary)

    return df
```
